### ① 印出最新文章的「作者」「標題」「時間」

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)
soup = BeautifulSoup(r.text)

df = pd.DataFrame(columns = ['作者', '標題', '時間'])
i = 0
for d in soup.find_all(class_="title"):
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text)
        data = r.find_all(class_='article-meta-value')
        df.loc[i] = [data[0].text, data[2].text, data[3].text]
        i += 1
    except:
        continue
    
df.head()
```

### ③ 試著爬爬看其他版的文章

```python
'''
Your Code
'''

url = 'https://www.ptt.cc/bbs/Option/index.html'
r = requests.get(url)
soup = BeautifulSoup(r.text)

for d in soup.find_all(class_='title'):
    print(d.text.replace('\t', '').replace('\n', ''))
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
        print('作者: ' + r.find_all(class_='article-meta-value')[0].text)
        print('標題: ' + r.find_all(class_='article-meta-value')[2].text)
        print('時間: ' + r.find_all(class_='article-meta-value')[3].text)
    except:
        continue
```
