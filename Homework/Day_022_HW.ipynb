```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome('./chromedriver')
browser.get('http://taqm.epa.gov.tw/taqm/tw/MonthlyAverage.aspx')
```

### ① 取出 台北市士林區 2018/01 – 2018/08 的 SO2 資料

```python
'''
Your Code
'''
selectYear = Select(browser.find_element_by_id('ctl05_ddlYear'))
selectYear.select_by_value('2018')

browser.find_element_by_id('ctl05_btnQuery').click()
```

```python
source = browser.page_source

soup = BeautifulSoup(source, 'html.parser')
table = soup.find('table', class_='TABLE_G')

data = {} #for dictionary
key = '' #set as indicators
date = '' #set as indicators
for item in table.find_all('td'):
    if 'style' in item.attrs:
        key = item.text #gas name
        data[key]={} #subdictionary for each gas
    elif 'class' not in item.attrs:
        if item.text.strip() != '': #if not blank
            if '/'in item.text: #cuz date has '/' 
                date = item.text
            else:
                data[key][date] = item.text

import pandas as pd
data_new = pd.DataFrame(data)
data_new
```
