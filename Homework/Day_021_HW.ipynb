## 1. 取出今天所有的新聞

```python
from bs4 import BeautifulSoup
from selenium import webdriver

browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get('https://www.ettoday.net/news/news-list.htm')

import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
```

```python
'''
Your Code
'''
d = soup.find(class_='part_list_2').find_all('h3')

import re

for title in d:
    date = title.find(class_ = 'date')
    subject = title.find('a')
    if re.compile('2020/03/04 [0-1][0-9]:[0-5][0-9]').match(date.text):
        print(date.text, subject.text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
'''
Your Code
'''
browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get('https://www.ettoday.net/news/news-list-2020-3-1-0.htm')

for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

d = soup.find('div', class_='part_list_2').find_all('h3')
```

```python
for title in d:
    date = title.find(class_='date')
    subject = title.find('a')
    if re.compile('2020/03/01 [1][5-7]:[0-5][0-9]').match(date.text):
        print(date.text, subject.text)
```
