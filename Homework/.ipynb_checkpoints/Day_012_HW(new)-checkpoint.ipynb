## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

### ① 取出今天所有的發文

```python
import re

today = '^2020/11/03 '

for d in soup.find(class_="part_list_2").find_all('h3'):
    if re.match(today, d.find(class_="date").text):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import numpy as np
import pandas as pd


date = []
tag = []
title = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    if re.match(today, d.find(class_="date").text):
        #print(d.find(class_="tag").text)
        date.append(d.find(class_="date").text)
        tag.append(d.find(class_="tag").text)
        title.append(d.find_all('a')[-1].text)
data = pd.DataFrame({"date": date, "tag": tag, "title": title})
data.head()
```

### ③ 哪一個類別的文章最多

```python
tags = data['tag'].unique()

for t in tags:
    print(t, data[data['tag']==t].shape[0])
```

體育類型的文章最多
