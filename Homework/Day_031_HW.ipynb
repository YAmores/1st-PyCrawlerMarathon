### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
"""
Your Code Here
"""
url = "https://news.chosun.com/international/"
headers ={"Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
          "Accept-Encoding": "br, gzip, deflate",
          "Host": "cdn.interworksmedia.co.kr",
          "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.0.5 Safari/605.1.15",
          "Accept-Language": "en-us",
          "Referer": "https://news.chosun.com/",
          "Connection": "keep-alive"}
r = requests.get(url, headers = headers)
response = r.text
response
```
