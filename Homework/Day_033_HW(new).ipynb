```python
url = 'https://github.com/new'

import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.text)
```

```python
res = requests.get(url,verify = False, cookies = {'logge_in': 'no'})
soup = BeautifulSoup(res.text,'html.parser')
print(soup.text) 
```
